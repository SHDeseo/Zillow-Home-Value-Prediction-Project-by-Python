```python
from matplotlib import pyplot
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
```

## Data Load

```python
train16 = pd.read_csv("train_2016_v2.csv", parse_dates=["transactiondate"])
print(train16.shape)
train16.tail()
```

```python
property16 = pd.read_csv("properties_2016.csv")
print(property16.shape)
property16.tail()
```

## join train & properties

```python
new_train16 = train16.set_index('parcelid').join(property16.set_index('parcelid'))
new_train16.head()
```

```python
total_prop = new_train16[["airconditioningtypeid", "heatingorsystemtypeid", "bathroomcnt", "calculatedbathnbr", "threequarterbathnbr", "fullbathcnt", "roomcnt", "bedroomcnt", "fireplacecnt", "fireplaceflag", "logerror"]]
print(total_prop.shape)
total_prop.tail()
```

## preprocessing & fill in missing value

### hvac (ëƒ‰ë‚œë°©)

```python
new_train16.loc[new_train16["heatingorsystemtypeid"] == 2.0, "airconditioningtypeid"] = 1.0
new_train16.loc[new_train16["airconditioningtypeid"] == 1.0, "heatingorsystemtypeid"] = 2.0

new_train16.loc[np.isnan(new_train16["airconditioningtypeid"]), "airconditioningtypeid"] = 0.0
new_train16.loc[np.isnan(new_train16["heatingorsystemtypeid"]), "heatingorsystemtypeid"] = 0.0
new_train16[["airconditioningtypeid", "heatingorsystemtypeid"]].isna().sum()
```

### fireplace (ë²½ë‚œë¡œ)

```python
new_train16["fireplaceflag"] = new_train16["fireplaceflag"].astype(float) # True/False -> 1.0/0.0
```

```python
new_train16.loc[~np.isnan(new_train16["fireplacecnt"]), "fireplacecnt"] = 1.0
new_train16.loc[~np.isnan(new_train16["fireplaceflag"]) , "fireplaceflag"] = 1.0
new_train16.loc[np.isnan(new_train16["fireplacecnt"]), "fireplacecnt"] = 0.0
new_train16.loc[np.isnan(new_train16["fireplaceflag"]), "fireplaceflag"] = 0.0

new_train16[["fireplacecnt", "fireplaceflag"]].isna().sum()
```

### room & bath & bed (ë°©, í™”ìž¥ì‹¤, ì¹¨ì‹¤)

```python
ls = new_train16[np.isnan(new_train16["fullbathcnt"])]["bathroomcnt"]
new_train16.loc[np.isnan(new_train16["fullbathcnt"]), "fullbathcnt"] = ls

new_train16[np.isnan(new_train16["fullbathcnt"])]
```

```python
new_train16["threequarterbathnbr"].value_counts()
```

```python
new_train16.loc[np.isnan(new_train16["threequarterbathnbr"]), "threequarterbathnbr"] = 1.0
new_train16[np.isnan(new_train16["threequarterbathnbr"])]
```

## OLS Modeling

### Basic

```python
dfX0 = new_train16[["airconditioningtypeid", "heatingorsystemtypeid", "bathroomcnt", "threequarterbathnbr", "fullbathcnt", "roomcnt", "bedroomcnt", "fireplacecnt", "fireplaceflag"]]
dfX = sm.add_constant(dfX0)
dfy = new_train16[["logerror"]]
df = pd.concat([dfX, dfy], axis=1)
```

```python
model = sm.OLS(dfy, dfX)
result = model.fit()
print(result.summary())
```

### Category & Scale
- fireplace 0/1 (ìžˆë‹¤ /ì—†ë‹¤ ì²˜ë¦¬)

```python
model2 = sm.OLS.from_formula("logerror ~ C(airconditioningtypeid) + C(heatingorsystemtypeid) + scale(bathroomcnt) + scale(threequarterbathnbr) + scale(fullbathcnt)\
                             + scale(roomcnt) + scale(bedroomcnt) +  fireplacecnt + fireplaceflag", data=df)
result2 = model2.fit()
print(result2.summary())
```

### Category & Scale
- hvac(ëƒ‰ë‚œë°©), fireplace(ë²½ë‚œë¡œ) ìžˆë‹¤/ì—†ë‹¤ ì²˜ë¦¬ (0/1)

```python
new_train16.loc[~np.isnan(new_train16["airconditioningtypeid"]), "airconditioningtypeid"] = 1.0
new_train16.loc[~np.isnan(new_train16["heatingorsystemtypeid"]), "heatingorsystemtypeid"] = 1.0
new_train16.loc[np.isnan(new_train16["airconditioningtypeid"]), "airconditioningtypeid"] = 0.0
new_train16.loc[np.isnan(new_train16["heatingorsystemtypeid"]), "heatingorsystemtypeid"] = 0.0

new_train16[["airconditioningtypeid", "heatingorsystemtypeid"]].isna().sum()
```

```python
model3 = sm.OLS.from_formula("logerror ~ airconditioningtypeid + heatingorsystemtypeid + scale(bathroomcnt) + scale(threequarterbathnbr) + scale(fullbathcnt)\
                             + scale(roomcnt) + scale(bedroomcnt) +  fireplacecnt + fireplaceflag", data=df)
result3 = model3.fit()
print(result3.summary())
```
